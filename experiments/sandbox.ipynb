```python
import os
import numpy as np
import matplotlib.pyplot as plt
import librosa

import matplotlib
import scipy
```

```python
filep = os.path.join("../data/Fur_Elise.wav")
y, sr = librosa.load(filep)

S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128,
                                    fmax=8000)
S_dB = librosa.power_to_db(S, ref=np.max)
img = librosa.display.specshow(S_dB, x_axis='time',
                         y_axis='mel', sr=sr,
                         fmax=8000)
```
